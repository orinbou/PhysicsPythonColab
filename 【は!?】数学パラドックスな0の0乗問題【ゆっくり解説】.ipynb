グラフ：y=x^x

```python
# prompt: 関数 y=x^x のグラフを描画してください

import numpy as np
import matplotlib.pyplot as plt

def plot_xx(xmin, xmax):
  x = np.arange(xmin, xmax, 0.01)
  y = x**x
  plt.plot(x, y)
  plt.xlabel("x")
  plt.ylabel("y = x^x")
  plt.title("Graph of y = x^x")
  plt.grid(True)
  plt.show()

plot_xx(0, 2)
```

```python
# prompt: 関数 y=x^x のyの値が最も小さくなる時のxの値を教えてさい
# https://mathwords.net/xnoxjou

import numpy as np

def f(x):
  return x**x

# Find the minimum value of f(x) within a specified range
x_values = np.linspace(0.1, 1.5, 1000)  # Create a range of x values
y_values = f(x_values)
min_y = np.min(y_values)
min_x = x_values[np.argmin(y_values)]

print(f"The minimum value of y = x^x is approximately {min_y} at x = {min_x}")
```

# 負と正の領域をつなげたグラフを描いてみる（※負の世界は整数のみ）

xのx乗のはなし (はじめよう数学 5)：https://www.amazon.co.jp/dp/453560844X/


```python
# prompt: 関数 y=x^x のグラフを(-4<=x<=2)までの範囲で描画してください。但しx<=0の領域はxは1.0刻みでx>=0の領域は0.1刻みでお願いします。
import numpy as np
import matplotlib.pyplot as plt

mx = np.arange(-4, 1, 1.00)
my = mx**mx
plt.plot(mx, my)

px = np.arange( 0, 2, 0.01)
py = px**px
plt.plot(px, py)

plt.xlabel("x")
plt.ylabel("y = x^x")
plt.title("Graph of y = x^x")
plt.grid(True)
plt.show()
```

```python
# prompt: 関数 y=x^x のグラフを(-4<=x<=2)までの範囲で描画してください。但しx<=0の領域はxは1.0刻みでx>=0の領域は0.1刻みでお願いします。また、負の領域ではグラフは非連続になると思うので、誤解のないよう、連続的な線で結ばず、点で表現して下さい
import numpy as np
import matplotlib.pyplot as plt

# For x <= 0, step is 1.0
# The range is -4 to 0 inclusive. np.arange(-4, 0.1, 1.0) correctly yields [-4, -3, -2, -1, 0]
x_neg_range = np.arange(-4, 0.1, 1.0)
y_neg_range = x_neg_range**x_neg_range

# For x >= 0, step is 0.1
# The range is 0 to 2 inclusive. np.arange(0, 2.1, 0.1) correctly yields [0.0, 0.1, ..., 2.0]
x_pos_range = np.arange(0, 2.1, 0.1)
y_pos_range = x_pos_range**x_pos_range

plt.figure(figsize=(10, 6))
# Plot negative x part, showing discrete points for integer values
plt.plot(x_neg_range, y_neg_range, 'o', label='x <= 0 (points only)') # Plot as points only

# Plot positive x part
plt.plot(x_pos_range, y_pos_range, label='x >= 0 (step 0.1)')

plt.xlabel("x")
plt.ylabel("y = x^x")
plt.title("Graph of y = x^x for -4 <= x <= 2")
plt.grid(True)
plt.legend()
plt.axhline(0, color='gray', linewidth=0.8) # Add x-axis
plt.axvline(0, color='gray', linewidth=0.8) # Add y-axis
plt.show()
```
